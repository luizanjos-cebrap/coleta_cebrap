# Pontos de Inundações
Aqui filtraremos, da tabela de ocorrências de 2024, quais são as ocorrências referentes a inundações.

# Dependências
Este notebook depende do arquivo resultante do notebook '../download_data/pontos_ocorrencias_2024'

```python
data_path= get_data_diretorio()
assets_path = os.path.join(
    data_path,
    'assets'
)
```

```python
gdf_ocorrencias = gpd.read_parquet(
    os.path.join(
        assets_path, 
        'ocorrencias_2024.parquet'
    )
)
```

```python
gdf_ocorrencias.sample(3)
```

# Filtragem
Iremos filtrar pelas ocorrências descritas como 'INUNDACAO' na coluna 'ocorrencia'

```python
gdf_ocorrencias['ocorrencia'].unique()
```

```python
gdf_inund = gdf_ocorrencias.loc[gdf_ocorrencias['ocorrencia']=='INUNDACAO']
```

# Conferir informações

```python
gdf_inund.shape
```

```python
gdf_inund.explore()
```

```python
print(f'''
Caso mais recente de ocorrência de inundação: {gdf_inund['data'].max()}
Caso mais antigo de ocorrência de inundação: {gdf_inund['data'].min()}
''')
```

```python
print(
    f'A subprefeitura com mais ocorrências de inundações é: {
        gdf_inund['nm_subpref'].mode()[0]
    }'
)
```

# Salvar GDF

```python
save_parquet_excel(
    gdf_inund,
    fname='pontos_inundacoes_2024',
    data_path= assets_path,
    data_subpath='areas_inundacao'
)
```
